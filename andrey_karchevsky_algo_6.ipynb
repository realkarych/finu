## Задача 1

Напишите функцию, которая параметром будет принимать путь к папке, а возвращать размер этой папки.

````python
import os

def folder_size(path):
    total = 0
    for dirpath, dirnames, filenames in os.walk(path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            try:
                total += os.path.getsize(fp)
            except (OSError, FileNotFoundError):
                pass
    return total
````

## Задача 2

Дан текстовый файл. В нем написан текст, разбитый на предложения. Перемешайте слова в каждом предложении текста и запишите полученный текст в новый файл.

````python
import re
import random

input_path = 'input.txt'
output_path = 'output.txt'
with open(input_path, 'r', encoding='utf-8') as f:
    text = f.read()
sentences = re.split(r'(?<=[.!?])\s+', text)
result = []
for sent in sentences:
    words = re.findall(r'\S+', sent)
    word_list = [w for w in words if re.search(r'[a-zA-Zа-яА-ЯёЁ]', w)]
    random.shuffle(word_list)
    result.append(' '.join(word_list))
with open(output_path, 'w', encoding='utf-8') as f:
    f.write(' '.join(result))
````

## Задача 3

В файле расположен русский текст. Откройте этот файл, преобразуйте текст в транслит и запишите в новый файл.

````python
translit_map = {'а':'a','б':'b','в':'v','г':'g','д':'d','е':'e','ё':'e','ж':'zh','з':'z','и':'i','й':'y','к':'k','л':'l','м':'m','н':'n','о':'o','п':'p','р':'r','с':'s','т':'t','у':'u','ф':'f','х':'h','ц':'ts','ч':'ch','ш':'sh','щ':'sch','ъ':'','ы':'y','ь':'','э':'e','ю':'yu','я':'ya'}
input_path = 'input_ru.txt'
output_path = 'output_translit.txt'
with open(input_path, 'r', encoding='utf-8') as f:
    text = f.read()
result = []
for c in text:
    if c.lower() in translit_map:
        result.append(translit_map[c.lower()] if c.islower() else translit_map[c.lower()].capitalize())
    else:
        result.append(c)
with open(output_path, 'w', encoding='utf-8') as f:
    f.write(''.join(result))
````

## Задача 4

Дан текстовый файл. Получите количество букв в нем (пробелы, знаки препинания и специальные символы не учитываются).

````python
input_path = 'input.txt'
with open(input_path, 'r', encoding='utf-8') as f:
    text = f.read()
count = sum(1 for c in text if c.isalpha())
print(count)
````

## Задача 6

Напишите программу, которая запрашивает у пользователя два значения. Если хотя бы одно из них не является числом, то программа будет соединять их в одну строку. В остальных случаях, если оба значения являются числами, программа будет суммировать их.

Примеры выполнения программы:
```
Первое значение: 10
Второе значение: 50
Результат: 60
```

```
Первое значение: hello
Второе значение: world
Результат: helloworld
```






````python
a = input("Первое значение: ")
b = input("Второе значение: ")
try:
    x = float(a)
    y = float(b)
    print("Результат:", x + y)
except ValueError:
    print("Результат:", a + b)
````
