## Разминка 1
Сгенерируем массив `arr` размера `(4, 7)` из `U(0, 20)` и выполним линейную нормализацию вида `a*x+b` так, чтобы `min=0.0`, `max=1.0`.

```python
import numpy as np

np.random.seed(42)  # фиксируем seed для воспроизводимости

arr = np.random.uniform(0, 20, size=(4, 7))

# Линейная нормализация в диапазон [0, 1]:
# x_norm = (x - x_min) / (x_max - x_min)
x_min, x_max = arr.min(), arr.max()
arr_norm = (arr - x_min) / (x_max - x_min)

arr, arr_norm.min(), arr_norm.max()
```

## Разминка 4
Решим матричное уравнение `A*X*B = -C`.

План:
1) Решаем `A*Y = -C` (получаем `Y`).
2) Решаем `X*B = Y` (получаем `X`).

Так избегаем явного вычисления обратных матриц.

```python
A = np.array([[-1, 2, 4],
              [-3, 1, 2],
              [-3, 0, 1]], dtype=float)

B = np.array([[3, -1],
              [2,  1]], dtype=float)

C = np.array([[7, 21],
              [11, 8],
              [8,  4]], dtype=float)

# 1) A*Y = -C  => Y = solve(A, -C)
Y = np.linalg.solve(A, -C)

# 2) X*B = Y  => (B^T * X^T = Y^T) => X^T = solve(B^T, Y^T)
X = np.linalg.solve(B.T, Y.T).T

X
```

```python
# Проверка: A*X*B должно быть равно -C (с учетом численной погрешности)
np.allclose(A @ X @ B, -C)
```

### 1) Загрузка данных
Считаем файл в `numpy`-массив типа `int32` с помощью `np.loadtxt` и выведем первые 5 строк.

```python
data = np.loadtxt(r"minutes_n_ingredients.csv", delimiter=",", skiprows=1, dtype=np.int32)

data[:5]
```

### 2) Среднее, минимум, максимум, медиана
Считаем статистики по каждому столбцу, кроме первого (`id`).

```python
minutes = data[:, 1]
n_ing = data[:, 2]

stats = {
    "minutes": {
        "mean": minutes.mean(),
        "min": minutes.min(),
        "max": minutes.max(),
        "median": np.median(minutes),
    },
    "n_ingredients": {
        "mean": n_ing.mean(),
        "min": n_ing.min(),
        "max": n_ing.max(),
        "median": np.median(n_ing),
    }
}
stats
```

### 3) Ограничение сверху по квантилю $q_{0.75}$
Найдем квантиль `0.75` по `minutes` и «обрежем» значения сверху этим порогом.

Чтобы порог оставался целым (как и минуты), берем метод `higher` — возвращает значение из выборки.

```python
q75 = np.quantile(minutes, 0.75, method="higher")

minutes_capped = np.minimum(minutes, q75).astype(np.int32)

q75, minutes.max(), minutes_capped.max()
```

```python
data_capped = data.copy()
data_capped[:, 1] = minutes_capped
data_capped[:5]
```

### 4) Нулевые длительности

```python
zero_mask = data_capped[:, 1] == 0
zero_count = int(zero_mask.sum())

data_fixed = data_capped.copy()
data_fixed[zero_mask, 1] = 1

zero_count, (data_fixed[:, 1] == 0).sum()
```

### 14) Группировка по длительности и формирование 3D массива
Группы:
- **короткие**: `< 10` минут
- **стандартные**: `10 <= minutes < 20` минут (включили 10, чтобы все значения попали в одну из групп)
- **длинные**: `>= 20` минут

Далее берём одинаковое число рецептов из каждой группы (минимум по размерам групп), чтобы можно было собрать единый 3D массив.

```python
m = data_with_simple[:, 1]

short = data_with_simple[m < 10]
standard = data_with_simple[(m >= 10) & (m < 20)]
long = data_with_simple[m >= 20]

short.shape, standard.shape, long.shape
```

```python
min_n = min(short.shape[0], standard.shape[0], long.shape[0])

arr_3d = np.stack([short[:min_n], standard[:min_n], long[:min_n]], axis=0)

arr_3d.shape
```
